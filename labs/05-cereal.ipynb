## Exercise 1: Protein Powder
Carbs, fats and proteins are the three primary macro nutrients. Create a figure plotting the distribution of each of these macro nutrients together (i.e., three distributions on a single plot). Make sure to provide a legend.

```python
# Enter all of your code for exercise 1 here. Feel free to add more cells if you need to:
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the distributions
plt.figure(figsize=(10, 6))
sns.kdeplot(cereal['carbo'], label='Carbohydrates', fill=True)
sns.kdeplot(cereal['protein'], label='Proteins', fill=True)
sns.kdeplot(cereal['fat'], label='Fats', fill=True)

# Add title and labels
plt.title('Distribution of Macronutrients in Cereal')
plt.xlabel('Amount')
plt.ylabel('Density')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

## Exercise 5: America Part 2
Make five plots comparing the relationships of carbo, sugars, calories, protein, and fat with rating.

```python
# Write your code to compare the various variables with rating below:
plt.figure(figsize=(15, 25))

plt.subplot(5, 1, 1)
sns.scatterplot(x='carbo', y='rating', data=cereal)
plt.title('Carbohydrates vs Rating')
plt.xlabel('Carbohydrates')
plt.ylabel('Rating')

plt.subplot(5, 1, 2)
sns.scatterplot(x='sugars', y='rating', data=cereal)
plt.title('Sugars vs Rating')
plt.xlabel('Sugars')
plt.ylabel('Rating')

plt.subplot(5, 1, 3)
sns.scatterplot(x='calories', y='rating', data=cereal)
plt.title('Calories vs Rating')
plt.xlabel('Calories')
plt.ylabel('Rating')

plt.subplot(5, 1, 4)
sns.scatterplot(x='protein', y='rating', data=cereal)
plt.title('Protein vs Rating')
plt.xlabel('Protein')
plt.ylabel('Rating')

plt.subplot(5, 1, 5)
sns.scatterplot(x='fat', y='rating', data=cereal)
plt.title('Fat vs Rating')
plt.xlabel('Fat')
plt.ylabel('Rating')

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

Of the variables carbo, sugars, calories, protein, and fat, which has the
strongest relationship with rating? Justify your answer.

I think that sugars have the strongest relationship to rating because as the sugar content in cereals increases, the calorie content also decreases significantly. The strong positive correlation suggests that sugar content is a significant contributor to the calorie content in cereals.

## Exercise 8: Captain Crunch the Numbers
Provide one additional insight from this dataset that you found interesting. Create at least one figure and explain why the figure was interesting to you.

```python
# Create the extra plot below:
import matplotlib.pyplot as plt
import seaborn as sns
cereal_type_counts = cereal['type'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=cereal_type_counts.index, y=cereal_type_counts.values)
plt.title('Distribution of Cereal Types')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='mfr', hue='type', data=cereal)
plt.title('Distribution of Manufacturer and Cereal Type')
plt.xlabel('Manufacturer')
plt.ylabel('Count')
plt.legend(title='Type')
plt.show()
```

There are more boxes of cold cereal sold when compared to hot cereal. The demand for hot cereal takes up about less than 10 percent.
